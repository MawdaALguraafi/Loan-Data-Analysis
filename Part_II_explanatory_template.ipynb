## Investigation Overview
لإhis project analyzes a dataset from Prosper, a peer-to-peer lending platform, which contains detailed records of over 113,000 loans issued between 2005 and 2014. Each entry in the dataset represents a unique loan and includes information about borrower profiles (such as employment status and income range), loan characteristics (including loan amount, interest rates, and loan terms), and loan outcomes (such as current, defaulted, or completed status).

The primary goal of this exploratory data analysis is to understand how different borrower and loan attributes influence borrowing costs and loan performance. By examining key variables and their relationships, we aim to uncover patterns that can help in assessing loan risk and borrower creditworthiness. Visualizations will be used to support insights and guide further analysis

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
df=pd.read_csv('clean_dataset')
df.head()
````

````python
df.drop(columns=["Unnamed: 0"], inplace=True)
````

## (Visualization 1)
 This chart shows the distribution of Borrower APR values across loans.
 Most values fall between 0.1 and 0.3, with some higher values that may be outliers.

````python
#Who Are the Borrowers?
sns.histplot(df['BorrowerAPR'], bins=30, kde=True, color='teal')
plt.title('Distribution of Borrower APR')
plt.xlabel('Borrower APR')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
````

## (Visualization 2)

This box plot illustrates how the number of investors varies with the borrower’s employment status. Self-employed and part-time borrowers tend to have more variability in investor count.

````python
order = ['HR', 'E', 'D', 'C', 'B', 'A', 'AA']
sns.boxplot(data=df, x='ProsperRating (Alpha)', y='BorrowerAPR', order=order)
plt.title('Borrower APR by Prosper Rating')
plt.xlabel('Prosper Rating (Alpha)')
plt.ylabel('Borrower APR')
plt.grid(True)
plt.show()
````

## (Visualization 3)

This grouped bar chart shows how loan outcomes (Completed, Defaulted, etc.) vary across employment statuses. Employed borrowers tend to have more completed loans, while unemployed borrowers show higher default rates.

````python
sns.scatterplot(data=df, x='DebtToIncomeRatio', y='BorrowerRate', alpha=0.5)
plt.title('Borrower Rate vs Debt-to-Income Ratio')
plt.xlabel('Debt-to-Income Ratio')
plt.ylabel('Borrower Rate')
plt.grid(True)
plt.show()
````
